```python
import csv 
import os
import cv2
import glob
import numpy as np
from shutil import copyfile
from shutil import move
```

```python
FOL_VIEWS = "/home/hthieu/AICityChallenge2019/data/Track2Data/image_train_by_view"
views = glob.glob(os.path.join(FOL_VIEWS,"*"))
views.sort()
imgs2id = {}
with open("../data/track2_train.csv") as fi:
    csv_reader = csv.reader(fi, delimiter=',')
    for info in csv_reader:
        imgs2id[info[1]] = info[0]
with open("../data/track2_train_best_imgs.csv", "w") as fo:
    for view in views:
        imgs = glob.glob(os.path.join(view,"*.jpg"))
        file_size = [(img,os.path.getsize(img)) for img in imgs]
        file_size.sort(key = lambda x: x[1], reverse = True)
        for x in file_size[:1]:
            file_name = os.path.basename(x[0])
            fo.write("{},{},{}\n".format(img2id[file_name],file_name,
                     int(os.path.dirname(x[0]).split("/")[-1]),
                    x[1]))
    fo.close()
    
```

```python
import os
import tensorflow as tf
from matplotlib import pyplot as plt
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
# The GPU id to use, usually either "0" or "1"
os.environ["CUDA_VISIBLE_DEVICES"]= ""
```

```python
img = plt.imread('/home/hthieu/AICityChallenge2019/data/Track2Data/image_test/000290.jpg')
img2= plt.imread('/home/hthieu/AICityChallenge2019/data/Track2Data/image_test/011478.jpg')
img = tf.convert_to_tensor(img)
img2= tf.convert_to_tensor(img2)
origin = img
img = tf.image.resize_images(img, (128,256))
img2= tf.image.resize_images(img2,(128,128))
img = tf.cast(img, tf.int32)
img2= tf.cast(img2, tf.int32)
with tf.Session() as sess:
    img,img2,origin = sess.run([img,img2,origin])
    plt.imshow(img)
    plt.figure()
    plt.imshow(origin)
    plt.figure()
    tmp = img[:,-128:,:]
    plt.imshow(tmp)
    plt.figure()
    plt.imshow(img2)
    
# reader = tf.WholeFileReader()
# key, value = reader.read(filenames)
```
